```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

# Zero-coupon bonds

```python
t = np.arange(0,31) # time horizon
r1 = 0.04 # low rate
r2 = 0.1  # high rate

f1 = np.exp(-r1*t)
f2 = np.exp(-r2*t)

#plot data
plt.figure(figsize=(8,5)) # set the figure size
plt.plot(t, f1, linewidth=2)
plt.plot(t, f2, linewidth=2)
plt.xlabel('time (year)')
plt.ylabel('p(t,T)')
plt.legend(labels =['low rate', 'high rate'], loc='best')
plt.xlim(0,30)
plt.ylim(0,1.2)
plt.show()
```
